```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("./data/train_clean.csv", encoding = "latin1")
train
```

```python
clarity = {
    "IF": 8,
    "VVS1": 7,
    "VVS2": 6,
    "VS1": 5,
    "VS2": 4,
    "SI1": 3,
    "SI2": 2,
    "I1": 1
}
```

```python
cut = {
    "Ideal": 5,
    "Premium": 4,
    "Very Good": 3,
    "Good": 2,
    "Fair": 1
}
```

```python
color = {
    "D": 7,
    "E": 6,
    "F": 5,
    "G": 4,
    "H": 3,
    "I": 2,
    "J": 1
}
```

```python
train["clarity"] = train.clarity.map(clarity)
train["cut"] = train.cut.map(cut)
train["color"] = train.color.map(color)
```

```python
train.head()
```

```python
train.drop("id", axis=1, inplace=True)
```

```python
train.corr()
```

```python
train.to_csv("./data/train_clean.csv")
```
